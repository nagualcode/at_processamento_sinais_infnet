```python

import numpy as np
import matplotlib.pyplot as plt


#questao 01 - c)

frequencia_amostragem = 8000


frequencia_sinal = 2


amplitude = 3


tempo = np.linspace(0, 1, 100)


sinal = amplitude * np.sin(2 * np.pi * frequencia_sinal * tempo)


plt.figure()

=
plt.plot(tempo, sinal)


plt.title('Sinal Senoidal')


plt.xlabel('Tempo (s)')


plt.ylabel('Amplitude')

indices_amostras = np.arange(len(tempo))

amostras = tempo[indices_amostras]


plt.scatter(amostras, sinal[indices_amostras], color='pink')


plt.show()
```

```python
# QUESTÃO 03:

import numpy as np
from scipy.stats import norm, uniform

# Gerar sinal X1 com distribuição normal
mean = 1
std = 0.1
X1 = norm.rvs(mean, std, size=1000)

# Gerar sinal X2 com distribuição uniforme
a, b = -1, 1
X2 = uniform.rvs(a, b-a, size=1000)

# Calcular relação sinal/ruído dos dois sinais
snr1 = np.mean(X1) / np.std(X1)
snr2 = np.mean(X2) / np.std(X2)

# Calcular coeficiente de variação dos dois sinais
cv1 = np.std(X1) / np.mean(X1)
cv2 = np.std(X2) / np.mean(X2)

print("Sinal Ruido X1:", snr1)
print("Sinal Ruido X2:", snr2)
print("Coeficiente de Variação X1:", cv1)
print("CCoeficiente de Variação X2:", cv2)
```

```python
#Questão 05

import matplotlib.pyplot as plt
import cv2
import numpy as np
import requests
from io import BytesIO

# Obtenha a imagem a partir da URL
url = "https://picsum.photos/800/800"
response = requests.get(url)

# Carregue a imagem como um array numpy
image = cv2.imdecode(np.asarray(bytearray(response.content), dtype="uint8"), cv2.IMREAD_COLOR)

# Divida a imagem em canais RGB
r, g, b = cv2.split(image)

# Desloque cada canal em uma posição diferente
r_shift = np.roll(r, 5, axis=1)
g_shift = np.roll(g, 50, axis=0)
b_shift = np.roll(b, 90, axis=0)

# Junte os canais deslocados novamente em uma imagem
image_shift = cv2.merge((r_shift, g_shift, b_shift))

# Mostre a imagem usando matplotlib
plt.imshow(image_shift)

```

```python
#QUESTÃO 06:

import matplotlib.pyplot as plt
import cv2
import numpy as np
import requests
from io import BytesIO

# Obtenha a imagem a partir da URL
url = "https://picsum.photos/800/800"
response = requests.get(url)

# Carregue a imagem como um array numpy
image = cv2.imdecode(np.asarray(bytearray(response.content), dtype="uint8"), cv2.IMREAD_COLOR)

# Divida a imagem em canais RGB
r, g, b = cv2.split(image)

# Exibe imagem e histogramas de cada cor RGB:

plt.subplot(311)
plt.hist(b.ravel(), 256, [0, 256], color='blue')
plt.grid()

plt.subplot(312)
plt.hist(g.ravel(), 256, [0, 256], color='green')
plt.grid()

plt.subplot(313)
plt.hist(r.ravel(), 256, [0, 256], color='red')

plt.grid()
plt.show()
plt.imshow(image)


```

```python
# Criar novos canais de cores com as intensidades desejadas
r_new = r // 2 # Intensidade do canal vermelho é a metade da original
b_new = b # Intensidade do canal azul é a mesma da original
g_new = g * 2 # Intensidade do canal verde é o dobro da original

# Juntar os novos canais de cores para criar a nova imagem
img_new = cv2.merge([b_new, g_new, r_new])

# Exibe imagem e histogramas de cada cor RGB:

plt.subplot(311)
plt.hist(b_new.ravel(), 256, [0, 256], color='blue')
plt.grid()

plt.subplot(312)
plt.hist(g_new.ravel(), 256, [0, 256], color='green')
plt.grid()

plt.subplot(313)
plt.hist(r_new.ravel(), 256, [0, 256], color='red')

plt.grid()
plt.show()
plt.imshow(img_new)

```
